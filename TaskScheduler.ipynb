```python
import google.generativeai as genai
import json
genai.configure(api_key="API KEY ")
model = genai.GenerativeModel("models/gemini-2.5-flash")
```

```python
system_prompt = """
You are an intelligent productivity and scheduling assistant.

Your task is to:
- create balanced daily schedules
- organize tasks efficiently
- avoid overload
- include breaks when necessary
- prioritize important tasks
- create realistic and productive plans

Always generate schedules in a clean time-table format.
"""
```

```python
print("Gemini Chatbot Started!")
print("Type 'exit' to stop.\n")
chat_history=[]
while True:
    user_input=input("You: ")
    if user_input.lower() =="exit":
        break
    chat_history.append(f"User: {user_input}")
    full_prompt=system_prompt+ "\n\n"+"\n".join(chat_history)
    response=model.generate_content(full_prompt)
    bot_reply=response.text
    print("Bot:",bot_reply)
    chat_history.append(f"Bot: {bot_reply}")
    
```

```python
wake_time=input("Wake up time: ")
sleep_time=input("Sleep time : ")
fixed_tasks=input("Enter fixed tasks (college/job/classes etc): ")
flexible_tasks=input("Enter flexible tasks(study/gym/coding etc):")
```

```python
user_schedule_prompt = f"""
Wake up time: {wake_time}

Sleep time: {sleep_time}

Fixed tasks:
{fixed_tasks}

Flexible tasks:
{flexible_tasks}

Create a balanced and productive daily schedule.
"""
```

```python
response=model.generate_content(
    system_prompt + "\n\n" + user_schedule_prompt
)
print(response.text)
```
